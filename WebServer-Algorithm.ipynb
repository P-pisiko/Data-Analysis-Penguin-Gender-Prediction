```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
pengu  = sns.load_dataset('penguins')
```

```python
pengu.drop("island",axis=1,inplace=True)
```

```python
pengu["species"].replace(["Adelie", "Chinstrap", "Gentoo"], [0, 1, 2], inplace=True)
pengu["sex"].replace(["Male", "Female"], [0, 1], inplace=True)
```

```python
pengu.dropna(inplace=True,axis=0)
```

```python
pengu
```

```python
X,y = pengu.drop("sex",axis=1), pengu["sex"].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=65)

from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression(max_iter=1500)
logmodel.fit(X_train, y_train)
```

<h3>Wep service part</h3>

```python
from flask import Flask, request, jsonify
app = Flask(__name__)
```

```python
aa = pengu.drop("sex",axis=1).iloc[0]
aa["species" ] = 1
aa
```

```python
@app.get("/penguin/sample")
def get_example():
    return pengu.drop("sex",axis=1).iloc[0].to_json(orient="index")
```

```python
@app.get("/penguin/explain")
def explane_sample():
    ornek = pengu.drop("sex",axis=1).iloc[0]
    ornek["species" ] = "Species are in the order of 0,1,2 representing Adelie,Chinstrap,Gentoo"
    ornek["sex" ] = "Sex is in the order of 0,1 representing Male,Female"
    ornek["bill_depth_mm" ] = "DVAR value between 13.1 and 21.5"
    ornek["bill_length_mm" ] = "DVAR value between 32.1 and 50.6"
    ornek["flipper_length_mm" ] = "DVAR value between 172.0 and 231.0"
    ornek["body_mass_g" ] = "DVAR value between 2700.0 and 6300.0"
    
    return ornek.to_json(orient="index") 
```

```python
@app.post("/penguin/evaluate")
def test_sample():
    if request.is_json:
        data = request.get_json()
        sample = pd.DataFrame([data])
        result = logmodel.predict(sample)
        if result[0] == 0:
            return jsonify({"sex": "Male"})
        else:
            return jsonify({"sex": "Female"})
        
    else:
        return jsonify({"error": "Request must be JSON"})
```

```python
if __name__ == "__main__":
    app.run()
```
